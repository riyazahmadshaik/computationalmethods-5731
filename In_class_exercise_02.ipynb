(1) Write a Python program to find the duplicate elements in a given array of integers. Return -1 If there are no such elements. (4 points)

```python
# write your answer here

def duplicate(d):
  dupli_element = set() 
  dictionary = {}

  for i in range(len(d)):
    if d[i] not in dictionary.keys():
      dictionary[d[i]] = 1
    else:
     dictionary[d[i]] += 1
  for j in dictionary.keys():
     if dictionary[j] > 1:
        dupli_element.add(j)
  if dupli_element:
    return dupli_element
  else:
    return -1
 
k = [7,7,4,5,6,6,4,3,3,9,9,4,2,1]
l = [5,6,7,9]

print(duplicate(k))
print(duplicate(l))
```

(2) Write a Python program to select all the Sundays of a specified year. (4 points)



```python
# write your answer here
from datetime import date, timedelta
def all_sundays(year):
  dt = date(year,1,1)
  dt += timedelta(days = 6 - dt.weekday())
  while dt.year == year:
    yield dt
    dt += timedelta(days = 7)

for s in all_sundays(2020):
  print(s)

```
